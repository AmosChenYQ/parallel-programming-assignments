```python
import matplotlib.pyplot as plt
import numpy as np
import math

from google.colab import drive
drive.mount('/gdrive')

dimensions = []
threads = []
results = []

with open('/gdrive/MyDrive/Colab Data/threads.log') as file:
  for line_with_num in list(enumerate(file)):
    # remove spaces at the end
    line_num = line_with_num[0]
    line = line_with_num[1].rstrip()
    # get the dimension of matrix with the count of threads used in calcuation
    print(line_with_num)
    if line_num % 6 == 1:
      dimension = int(line.split(' ')[1])
      thread = int(line.split(' ')[3])
      dimensions.append(dimension)
      threads.append(thread)
    if line_num % 6 == 5:
      results.append(math.log(float(line.split(' ')[-1][:-1])))

# print(dimensions)
# print(threads)
# print(results)



ax = plt.axes(projection='3d')
# ax.view_init(60, 35)
ax.set_xlim3d(1, 16)
ax.set_ylim3d(1, 8)
ax.invert_xaxis()
ax.scatter3D(dimensions, threads, results, c=results)
ax.set_title('Plot of total time(in log) versus number of threads and number of rows of matrix')
```
